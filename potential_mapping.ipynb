```python
import re

# Step 1: Read the residues, states, and potential values from Pt_mbeta.csv
residue = []
statex = []
pot = []

with open('Pt_mbeta.csv', 'r') as f3:
    file3 = f3.readlines()
    for line in file3:
        se1 = re.search(r'^(...)\,\s(\d)\,\s(\d+\.\d+)', line)
        if se1:
            residue.append(se1.group(1))
            statex.append(int(se1.group(2)))
            pot.append(float(se1.group(3)))

# Step 2: Read the list of filenames from ALPHA_100x.txt
with open('fam_beta.txt', 'r') as f1:
    file1 = f1.readlines()
    filenames = [line.strip() for line in file1]

# Step 3: For each filename, read the corresponding file and match residues and states
for file_name in filenames:
    print(file_name)
    resi = []
    state = []

    with open(f'C:\\Users\\chaud\\Music\\Project\\Project_timeseries\\Timeseries_return\\Beta\\{file_name}_state.csv', 'r') as f2:
        file2 = f2.readlines()
        for line in file2:
            se2 = re.search(r'^(...)\,\s(\d+)', line)
            if se2:
                resi.append(se2.group(1))
                state.append(int(se2.group(2)))

    # Step 4: Write the matches to new files
    with open(f'C:\\Users\\chaud\\Music\\Project\\Project_timeseries\\Timeseries_return\\Beta_mapped\\{file_name}_mapped.csv', 'w') as g:
        g.write("Residue, State, Potential\n")
        for i in range(len(resi)):
            for j in range(len(residue)):
                if resi[i] == residue[j] and state[i] == statex[j]:
                    g.write(f"{resi[i]}, {state[i]}, {pot[j]}\n")
print("done!!!!!!!!!!!!!!!")
```
